Enumerate

```python
# Enumerate bir listede index ve valueyi gösterir.enumerate() while iterating through the elements of a list, gives you both the index and the element itself.
```

```python
for element in enumerate(my_list):
    print(element)
```

```python
for (ix,value) in enumerate(my_list):
     print(value)
```

Zip

```python
foodList=["apple", "banana", "grape"]
caloriesList=[100, 150,70]
dayList=["monday","tuesday","wednesday"]
zipList=list(zip(foodList,caloriesList,dayList))
```
