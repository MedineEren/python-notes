### Inheritance

```python
class Musician():
    def __init__(self, name):
        self.name=name
        print("Musician class")
    def test1(self):
        print("test1")
    def test2(self):
        print("test2")
    
```

```python
ati=musician("ati242")
```

```python
ati.name
```

```python
ati.test1()
```

```python
class MusicianPlus(Musician):
    
    def __init__(self,name):
        Musician.__init__(self,name)
        print("musician plus")
        
    def test3(self):
        print("test3")
    
    #override
    def test1(self):
        print("test1 test1 test1")
```

```python
atilla=musicianPlus("atilla serin")
```

```python
atilla.name
```

```python
atilla.test1()
```

```python
atilla.test3()
```

```python
class student():
    def __init__(self,name,grade):
        self.name= name
        self.__grade=grade  #private variable
    def info(self):
        print(f"{self.name} grade is:{self.__grade}")
    def changesGrade(self,grade):
        self.__grade=grade
```

### ABSTRACTION

```python
from abc import ABC, abstractmethod
class animal(ABC):
    @abstractmethod
    def soundmake(self):
        pass
class Dog(animal):
    def soundmake(self):
        return "haw haw haw"
class Cat (animal):
    def soundmake(self):
        return "meow meow"
```

```python
dog=Dog()
```

```python
cat=Cat()
```

```python
print(dog.soundmake())
```

```python
print(cat.soundmake())
```

```python
from abc import ABC, abstractmethod
class car(ABC):
    @abstractmethod
    def maxSpeed(ABC):
        pass
```

```python
class Volvo(car):
    def maxSpeed(self):
        print("240")
```

```python
class Audi(car):
    def maxSpeed(self):
        print("260")
```

### __str__

```python
#__str__() fonksiyonu, nesneyi print() ettiğinde ekranda nasıl görüneceğini belirler.
```

```python
class fruit():
    def __init__(self, name,calories):
        self.name=name
        self.calories=calories

    def __str__(self):
        return f"{self.name}:{self.calories} calories"
    def __len__(self):
        return self.calories
```

```python
myFruit= fruit("waterMelon", "200")
```

```python
myFruit.calories
```

```python
myFruit.name
```

```python
print(myFruit)
```
