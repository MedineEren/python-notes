Args, kwargs(arguments ,key word arguments) 

```python
#birdden fazla girdi olduğunda kullanıla bilşr ya da inputun kaç tane olacağını bilmiyoruz
```

```python
def args_exmp(*args):
    print(args)
    return args

args_exmp(56, 97, 37, 64, 85, 79)
```

```python
def kwargs_exmp(**kwargs):
    print(kwargs)
kwargs_exmp(apple=100, banana=200, grape=70)
```

```python
def kwargs_exmp2(**kwargs):
    if "apple" in kwargs:
        print("appleeeee")
    else:
        print(":(")
kwargs_exmp2(apple=100, banana=200, grape=70)
    
```
